# Data Cleaning




 Handling Missing Values

```python
print(df.isnull().sum())
```

```python
df.drop('stays_in_weeks_nights', axis=1, inplace=True)    #reason: all values are NaN
```

```python
df.drop('reservation_status_date', axis=1, inplace=True)  #reason: the values are "post-event"
```

```python
df.drop('reservation_status', axis=1, inplace=True)       #reason: the values are highly colerated to the column is_canceled
```

```python
#Checking the number of reserved_room_type equal to assigned_room_type

mask_equal = df['reserved_room_type'] == df['assigned_room_type']
count_equal = mask_equal.sum()
print(count_equal)
```

```python
#Setting the NaN values from reserved_room_type with the values of assigned_room_type

mask = df['reserved_room_type'].isna()
df.loc[mask, 'reserved_room_type'] = df.loc[mask, 'assigned_room_type']
```

```python
#Droping the column 'company'

df['company'].isna().sum() 
df.drop('company', axis=1, inplace=True)   #reason: more than 90% of the values are NaN
```

```python
#Setting NaN values of 'stays_in_weekend_nights' as value 0

df['stays_in_weekend_nights'] = df['stays_in_weekend_nights'].fillna(0)
df['stays_in_weekend_nights'] = df['stays_in_weekend_nights'].astype(int)
```

```python
df['country'] = df['country'].fillna('Unknown')
```

```python
#Setting the NaN values from 'agent' to 0 and changing the data type to int

df['agent'] = df['agent'].fillna(0)
df['agent'] = df['agent'].astype(int)
```

```python
#Replace missing values in 'adr' with the median to handle outliers and preserve data

median_adr = df['adr'].median()
df['adr'] = df['adr'].fillna(median_adr)
```

```python
print(df.isnull().sum())
```

Spot and Treat Outliers

```python
df.boxplot()
plt.xticks(rotation=45)  # for better clarity
plt.show()
```

```python
#Checking the outliers at the column adr

df['adr'].plot.box()
plt.title('Boxplot for ADR')
plt.ylabel('ADR')
plt.show()
```

```python
#Calculate limits to find outliers in 'adr'

Q1 = df['adr'].quantile(0.25)
Q3 = df['adr'].quantile(0.75)
IQR = Q3 - Q1

lower_bound = Q1 - 1.5 * IQR    
upper_bound = Q3 + 1.5 * IQR

df = df[(df['adr'] >= lower_bound) & (df['adr'] <= upper_bound)]   #Keep only the data within these limits
```

```python
df['adr'].plot.box()
plt.show()
```

```python
#Checking outliers at column 'days_in_waiting_list'

df['days_in_waiting_list'].plot.box()
plt.title('Boxplot for waiting')
plt.ylabel('diwl')
plt.show()
```

```python
#Checking how many entries in the 'days_in_waiting_list' column exceed the threshold and how many do not

threshold = 61

num_very_large = (df['days_in_waiting_list'] > threshold).sum()
num_normal = (df['days_in_waiting_list'] <= threshold).sum()

print(f"Number of very large values (> {threshold}): {num_very_large}")
print(f"Number of normal values (<= {threshold}): {num_normal}")
```

```python
df = df[df['days_in_waiting_list'] <= 61].copy()  #keeping only the values under 61 because it is uncommon to be for more than 2 months on waiting list
```

```python
df_nonzero = df[df['days_in_waiting_list'] > 0]
df_nonzero['days_in_waiting_list'].plot.box()
plt.title('Boxplot for days_in_waiting_list > 0')
plt.show()
```

```python
#Checking outliers at column lead_time

df['lead_time'].plot.box()
plt.title('Boxplot for lead time')
plt.ylabel('diwl')
plt.show()
```

```python
#Calculate limits to find outliers in 'lead_time'

Q1 = df['lead_time'].quantile(0.25)
Q3 = df['lead_time'].quantile(0.75)
IQR = Q3 - Q1

lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

outliers = df[(df['lead_time'] < lower_bound) | (df['lead_time'] > upper_bound)]
print(f"{len(outliers)} outliers found")
```

```python
df = df[(df['lead_time'] >= lower_bound) & (df['lead_time'] <= upper_bound)]   #Keep only the data within these limits
```

```python
df = pd.get_dummies(df, columns=['meal'], drop_first=False)  #One-Hot Encoding
```

```python
#Label Encoding
seasons = {
    'December': 'Winter', 'January': 'Winter', 'February': 'Winter',
    'March': 'Spring', 'April': 'Spring', 'May': 'Spring',
    'June': 'Summer', 'July': 'Summer', 'August': 'Summer',
    'September': 'Autumn', 'October': 'Autumn', 'November': 'Autumn'
}


df['season'] = df['arrival_date_month'].map(seasons)

```

```python
# How many times a country appears
country_counts = df['country'].value_counts()

rare_countries = country_counts[country_counts < 50]
print(f"Number of rare countries: {len(rare_countries)}")
```

```python
# Rare countries
rare_country_list = country_counts[country_counts < 50].index

# Change them with "Other"
df['country'] = df['country'].apply(lambda x: 'Other' if x in rare_country_list else x)
```

```python
df = pd.get_dummies(df, columns=['country'], drop_first=True)  #One-Hot Encoding
```
